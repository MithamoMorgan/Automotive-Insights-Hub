```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = 'https://www.kaiandkaro.com/vehicles?availability=available&availability=sell_on_behalf'
total_pages = 59

car_list = []
    
for page_num in range(1, total_pages + 1):
    url = f'{base_url}&page={page_num}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    car_details = soup.find_all('div', class_ = 'chakra-linkbox css-1kh7oap')
    for car_detail in car_details:
        make_model = car_detail.find('h2', class_ = 'chakra-heading css-18j379d').text.strip()
        transmission_type = car_detail.find('span', class_ = 'css-evl6jo').text.strip()
        year = car_detail.find('span', class_ = 'chakra-badge css-1dub5x4').text.strip()
        cc = car_detail.find_all('span', class_ = 'css-evl6jo')[1].text.strip()
        usage_origin = car_detail.find('span', class_ = 'css-b03jaa').text.strip()
        price =car_detail.find('p', class_ = 'chakra-text css-0').text.strip()

        car_dict = {'Model':make_model,'Transimission_Type':transmission_type,'Year':year,'CC':cc,'Usage Origin':usage_origin,'Price':price}
        car_list.append(car_dict)

car_df = pd.DataFrame(car_list)
car_df
```

```python
# saving the data in a csv file
car_df.to_csv('Kai_Karo_cars.csv', index = False)
```
